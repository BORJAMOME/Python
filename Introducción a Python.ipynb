### Operadores Condicionales

#### IF

Se utiliza para decisiones de verdadero o falso

if condición1: instrucción1

else: instrucción2

```python
edad_1= 15

if edad_1 >= 18:
  print('Mayor de edad')

print('A esperar mayoria de edad')
```

```python
edad_2= int(input('Ingrese su edad: '))

edad_2

if edad_2>= 18:
  print('Persona mayor de 18 años en adelatante')

else:
  print('Persona menor de 18 años')
```

#### IF - ELIF - ELSE

Consite que evalua por bloque hasta se cumpla una condición

```python
edad_3= int(input("Ingrese la edad de la persona: "))

if edad_3<= 2:
  print('Por el rango de la edad, la persona es Bebe')

elif edad_3>= 3 and edad_3<= 13:
  print('Niño ')

elif edad_3>= 14 and edad_3<= 18:
  print("Adolescente")

elif edad_3>= 19 and edad_3<= 29:
  print("Joven")


elif edad_3>= 30 and edad_3<= 59:
  print("Adulto")


else:

  print('Adulto Viejoven')
```

```python
ventas= pd.read_csv('Ventas.csv')
ventas
```

### Cálculo de columna "Ingreso"

```python
ventas['Ingreso']= ventas['Unidades'] * ventas['Precio']
ventas
```

```python
ventas['Ingreso'].sum()
```

```python
ventas['Ingreso'].max()
```

```python
ventas['Ingreso'].min()
```

#### Columna Calculada 'Costo' 60%

```python
ventas['Costo'] = ventas["Ingreso"]*0.60
ventas
```

#### Columna Calculada Beneficio = Ingreso - Costo

```python
ventas['Beneficio']= ventas['Ingreso']-ventas['Costo']
ventas
```

#### Columna Calculada Comision - 5%

```python
ventas['Comision']= ventas['Ingreso']*-0.05
ventas
```

#### Columna calculada utilizando Funcion Sum axis=1

```python
ventas['Ingreso Neto']= ventas[['Ingreso','Comision']].sum(axis=1)
ventas
```

```python
# Quitar la columna 'Ingreso'

ventas.drop(['Ingreso'],axis=1) 
```

```python
#### Eliminar columnas 

ventas.drop(['Unidades','Ingreso','Precio','Costo','Beneficio','Comision'],axis=1)
```
