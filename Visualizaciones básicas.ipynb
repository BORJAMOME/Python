# Librerías principales para análisis y visualización


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Gráfico de pastel

```python
# Datos de ejemplo

etiquetas = ["manzana", "plátano", "naranja", "fresa", "kiwi"]
tamaño = [180, 200, 150, 15, 75]

#Generar visualización de Pie

plt.pie(tamaño, labels=etiquetas,autopct='%1.1f%%')

# Ejecutar visual
plt.show()
```

```python
# Colores modernos (paleta Flat UI)
colores = ["#e74c3c",  # rojo (manzana)
           "#f1c40f",  # amarillo (plátano)
           "#e67e22",  # naranja (naranja)
           "#ff5e78",  # rosa (fresa)
           "#2ecc71"]  # verde (kiwi)

# Generar visualización de Pie
plt.figure(figsize=(6,6))  # tamaño más limpio
plt.pie(tamaño, labels=etiquetas, autopct='%1.1f%%', colors=colores, startangle=140, wedgeprops={'edgecolor': 'white'})

# Hacer que el gráfico sea circular
plt.axis('equal')

# Mostrar visual
plt.title("Distribución de Tamaños de Frutas", fontsize=14)
plt.show()
```

# Gráfico de Histograma

```python
# Datos de ejemplo
edades = [
    23, 45, 34, 29, 60, 38, 41, 52, 67, 31,
    25, 48, 36, 33, 59, 71, 28, 26, 40, 64,
    37, 49, 55, 30, 44, 43, 39, 61, 22, 46,
    35, 32, 58, 50, 27, 24, 66, 70, 53, 42,
    21, 62, 19, 56, 63, 47, 20, 51, 57, 54,
    65, 60, 38, 33, 59, 31, 29, 26, 34, 36,
    48, 45, 41, 39, 44, 67, 28, 40, 30, 22,
    55, 43, 32, 46, 61, 27, 24, 50, 37, 21,
    19, 23, 35, 53, 20, 64, 52, 66, 62, 25,
    70, 71, 18, 18, 60, 29, 38, 26, 31, 58
]

# Generar visualización
plt.hist(edades)

# Mostrar visualización
plt.show()
```

```python
# Visualización optimizada
plt.figure(figsize=(10, 6))
plt.hist(edades, bins=10, color="#3498db", edgecolor="white", alpha=0.85)

# Títulos y etiquetas
plt.title("Distribución de Edades - Muestra de Madrid", fontsize=14, weight='bold')
plt.xlabel("Edad", fontsize=12)
plt.ylabel("Número de personas", fontsize=12)

# Estética del grid
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Mostrar visual
plt.tight_layout()
plt.show()
```

# Gráfico de barras con diferentes colores

```python
# Datos de ejemplo
x = ["A", "B", "C", "D"]
y = [23, 47, 35, 61]

# Generar visualización
sns.barplot(x=x,y=y,palette= 'Set2')

# Ejecutar la visualización
plt.show
```

```python
# Crear figura
plt.figure(figsize=(8, 5))

# Visualización moderna con Seaborn
sns.barplot(x=x, y=y, palette='Set2')

# Títulos y etiquetas
plt.title("Valores por Categoría", fontsize=14, weight='bold')
plt.xlabel("Categoría", fontsize=12)
plt.ylabel("Valor", fontsize=12)

# Mostrar visual
plt.tight_layout()
plt.show()
```

# Gráfico de conteo

```python
# Datos

frutas = ["manzana", "naranja", "plátano", "plátano", "kiwi", "manzana", 
          "naranja", "naranja", "fresa", "plátano", "naranja", "kiwi"]

# Generar visual

sns.countplot(frutas)

# Ejecutar visualización
plt.show()
```

```python
# Configurar el tamaño del gráfico
plt.figure(figsize=(8, 5))

# Generar visualización con colores modernos
sns.countplot(x=frutas, palette='Set2')

# Añadir título y etiquetas
plt.title('Frecuencia de Frutas', fontsize=16, weight='bold')
plt.xlabel('Fruta', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Rotar etiquetas del eje X si es necesario
plt.xticks(rotation=45, ha='right')

# Mostrar la visualización
plt.tight_layout()
plt.show()
```

# Gráfico de mapa de calor

```python
# Datos 

datos5 = np.random.rand(10,5)

# Generar visual

sns.heatmap(datos5, annot=True)

# Ejecutar visualización

plt.show()
```

```python
# Datos
np.random.seed(42)
datos5 = np.random.rand(10, 5)

# Tamaño del gráfico
plt.figure(figsize=(10, 6))

# Estilo general
sns.set(style="white")

# Heatmap con paleta moderna y anotaciones optimizadas
sns.heatmap(
    datos5,
    annot=True,
    fmt=".2f",
    cmap="YlOrBr",  # Paleta moderna y elegante
    linewidths=0.1,
    linecolor='white',
    square=True,
    cbar_kws={'label': 'Intensidad'},
    annot_kws={"size": 10, "weight": "bold", "color": "#222"}
)

# Título y ejes
plt.title("Mapa de Calor Moderno", fontsize=18, weight='bold', pad=15)
plt.xlabel("Columnas", fontsize=13)
plt.ylabel("Filas", fontsize=13)

# Quitar bordes molestos
sns.despine(left=False, bottom=False)

# Ajustar márgenes
plt.tight_layout()

# Mostrar
plt.show()
```
