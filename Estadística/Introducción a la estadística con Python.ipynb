# 1. Estadistica descriptiva

La **estadística descriptiva** se refiere a un conjunto de técnicas utilizadas para **organizar, resumir y presentar** datos de manera que sea más fácil interpretarlos. Su propósito es ofrecer una visión general de las características principales de un conjunto de datos, sin hacer inferencias o conclusiones más allá de los datos mismos.

### Componentes clave de la estadística descriptiva:

1. **Medidas de tendencia central**:
   - **Media**: Promedio de los datos.
   - **Mediana**: Valor central cuando los datos están ordenados.
   - **Moda**: Valor que más se repite.

2. **Medidas de dispersión**:
   - **Desviación estándar**: Mide cuánto se dispersan los datos respecto a la media.
   - **Rango**: Diferencia entre el valor máximo y el mínimo.
   - **Varianza**: Cuánto varían los datos en promedio respecto a la media.

3. **Medidas de forma**:
   - **Asimetría (skewness)**: Indica si los datos están sesgados hacia la izquierda o derecha.
   - **Curtosis**: Mide la "altitud" y "ancho" de la distribución de los datos.

4. **Tablas y gráficos**:
   - **Tablas de frecuencia**: Muestran cuántas veces ocurre cada valor o rango de valores.
   - **Gráficos**: Como histogramas, diagramas de cajas, etc., para visualizar la distribución de los datos.

### En resumen:
La estadística descriptiva **resume** los datos y te ayuda a obtener una visión general rápida de su comportamiento sin realizar inferencias sobre la población de la cual provienen esos datos.

# 2. Estadística inferencial

La **estadística inferencial** se ocupa de hacer **inferencias o predicciones** sobre una población a partir de una muestra de datos. A diferencia de la estadística descriptiva, que simplemente resume y organiza los datos, la inferencial se utiliza para tomar decisiones o hacer predicciones basadas en los resultados obtenidos a partir de una muestra representativa.

### Componentes clave de la estadística inferencial:

1. **Estimación**:
   - **Estimación puntual**: Usar un solo valor (como la media de una muestra) para estimar un parámetro poblacional.
   - **Intervalos de confianza**: Rango de valores dentro del cual se espera que se encuentre el parámetro poblacional con un cierto nivel de confianza.

2. **Pruebas de hipótesis**:
   - **Hipótesis nula (H₀)**: Suposición inicial de que no hay efecto o relación.
   - **Hipótesis alternativa (H₁)**: Suposición que se quiere probar, indicando que sí hay un efecto o relación.
   - **Valor p**: Probabilidad de obtener resultados más extremos que los observados, bajo la hipótesis nula. Si el valor p es bajo, se rechaza la hipótesis nula.

3. **Regresión**:
   - Se usa para modelar la relación entre una variable dependiente y una o más variables independientes, permitiendo predecir el comportamiento de la variable dependiente.

4. **Análisis de varianza (ANOVA)**:
   - Permite comparar medias entre tres o más grupos para determinar si al menos uno de ellos es significativamente diferente de los demás.

5. **Distribuciones de probabilidad**:
   - Uso de distribuciones como la normal, t de Student, chi-cuadrada, entre otras, para modelar datos y realizar inferencias.

### En resumen:
La **estadística inferencial** va más allá de los datos disponibles para **hacer afirmaciones** sobre una población mayor, utilizando técnicas matemáticas y probabilísticas para generalizar los resultados obtenidos a partir de una muestra.

```python
# Separar variables predictoras (X) y variable objetivo (y) usando data_latin

# Opción 1: por posición (todas menos la última como X, y la última como y)
X = data_latin[['danceability']].values       # Predictor: danceability
y = data_latin['track_popularity'].values     # Objetivo: track_popularity
```

```python
#Empezamos con nuestro modelo

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
#Entrenamiento del modelo

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Recuperamos la interseccion

print(regressor.intercept_)
```

Cuando el predictor (danceability) vale 0, el modelo predice un track_popularity de aproximadamente 39.83. Es el punto donde la línea de regresión cruza el eje vertical.

```python
#Recuperamos la pendiente

print(regressor.coef_)
```

Por cada unidad extra en tu predictor (danceability), el modelo aumenta la predicción de track_popularity en aprox. 10.07 puntos.

```python
#Calculamos la prediccion

y_pred = regressor.predict(X_test)
y_pred
```

```python
#Convermos el array en un df de pandas

df = pd.DataFrame({'Actual': y_test , 'Prediccion': y_pred})
df
```

```python
#Evaluamos el modelo automatizada, para comparar multiple modelos de forma simultanea
from sklearn.metrics import mean_squared_error
modelo = ['Regresion lineal']

for i, model in enumerate ([regressor]):
    y_train_pred = regressor.predict(X_train)
    y_test_pred = regressor.predict (X_test)

    print(f'Modelo benchmark: {modelo[i]}')

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f'Raiz del error cuadratico medio en train: {rmse_train}')
    print(f'Raiz del error cuadratico medio en test: {rmse_test}')
```

Los valores 25.46 (train) y 25.16 (test) indican el promedio del error de predicción, es decir, cuánto se desvía en promedio el modelo al predecir la variable y respecto al valor real.

```python
#R2
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print('r2 score for a worse model is', r2)
```

Esto indica que el modelo no está capturando ninguna relación útil entre tus X e Y, y lo hace peor que un modelo que simplemente predijera el promedio de y para todos los casos.
